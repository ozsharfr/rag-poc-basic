```python
def extract_sentence_rows(search_results):
    # קודם serialize
    if hasattr(search_results, "model_dump"):
        resp = search_results.model_dump()
    else:
        resp = search_results.dict()

    resources = resp.get("resources", {}) or {}
    sent = (resp.get("sentences") or {}).get("results", []) or []

    rows = []
    for r in sent:
        rid = r["rid"]
        meta = resources.get(rid, {})
        pos = r.get("position") or {}
        rows.append({
            "rid": rid,
            "title": meta.get("title"),
            "slug": meta.get("slug"),
            "score": r.get("score"),
            "text": r.get("text"),
            "page_number": pos.get("page_number"),
            "start": pos.get("start"),
            "end": pos.get("end"),
            "field": r.get("field"),
        })
    return rows
```

```python
import json

def extract_hits(resp: dict, mode: str = "paragraphs", top_k: int | None = 20):
    """
    mode: "paragraphs" או "sentences"
    מחזיר רשימת hits, כל אחד כולל:
    rid, title, slug, score, text, page_number, start, end, field, icon, created, modified, language
    """
    resources = resp.get("resources", {}) or {}
    results = ((resp.get(mode) or {}).get("results") or [])

    if top_k is not None:
        results = results[:top_k]

    out = []
    for r in results:
        rid = r.get("rid")
        meta = resources.get(rid, {})  # join לפי rid

        pos = r.get("position") or {}
        out.append({
            "rid": rid,
            "title": meta.get("title"),
            "slug": meta.get("slug"),
            "icon": meta.get("icon"),
            "created": meta.get("created"),
            "modified": meta.get("modified"),
            "kb_language": (meta.get("metadata") or {}).get("language"),
            "status": (meta.get("metadata") or {}).get("status"),

            "score": r.get("score"),
            "text": r.get("text"),

            "page_number": pos.get("page_number"),
            "start": pos.get("start"),
            "end": pos.get("end"),
            "field": r.get("field"),
            "field_type": r.get("field_type"),
        })
    return out

def to_dict(obj):
    # Pydantic v2
    if hasattr(obj, "model_dump"):
        return obj.model_dump()
    # Pydantic v1
    if hasattr(obj, "dict"):
        return obj.dict()
    # fallback (אם זה כבר dict)
    return obj
```

```python
import pdfplumber
import os
from bidi.algorithm import get_display

# הגדרות נתיבים
input_dir = r"C:\Users\ozsha\aa\pdf_downloads"
files_fine = []
d = {}

def fix_hebrew_bidi(text):
    if not text:
        return ""
    # הפונקציה get_display הופכת את העברית הוויזואלית ללוגית 
    # תוך שמירה על מספרים ואנגלית בכיוון הנכון
    return get_display(text)

files = os.listdir(input_dir)

for file in files:
    if not file.lower().endswith('.pdf'):
        continue
        
    file_path = os.path.join(input_dir, file)
    
    try:
        with pdfplumber.open(file_path) as pdf:
            text_pages = []
            for page in pdf.pages:
                raw_page_text = page.extract_text()
                if raw_page_text:
                    # תיקון ה-Bidi לכל עמוד
                    fixed_page = fix_hebrew_bidi(raw_page_text)
                    text_pages.append(fixed_page)
            
            full_text = "\n".join(text_pages)
            
            # בדיקה אם מילת המפתח קיימת בטקסט המתוקן
            if 'עובדות' in full_text:
                print(f"נמצא קובץ מתאים: {file}")
                files_fine.append(file)
                d[file] = full_text
                
    except Exception as e:
        print(f"שגיאה בעיבוד הקובץ {file}: {e}")

# עכשיו המילון 'd' מכיל טקסט קריא עם מספרים תקינים
```

```python
def extract_results(resources):

    results = []

    for res in resources:   # FindResource
        title = res.title

        for field in res.fields.values():
            for para in field.paragraphs.values():

                results.append({
                    "TEXT": para.text,
                    "SCORE": para.score,
                    "TITLE": title
                })

    return results

extract_results(response_find.resources.values())
```
